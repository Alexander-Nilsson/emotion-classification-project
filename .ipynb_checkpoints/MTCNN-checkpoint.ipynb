```python
###Simple CNN model###

class Face_Detection(nn.Module):
    def __init__(self, num_boxes=10):
        super(Face_Detection, self).__init__()

        self.num_boxes = num_boxes
        #Convolutional layers
        self.conv1 = nn.Conv2d(3, 32, kernel_size=3)
        self.pool1 = nn.MaxPool2d(2, 2)
        self.conv2 = nn.Conv2d(32, 64, kernel_size=3)
        self.pool2 = nn.MaxPool2d(2, 2)
        self.conv3 = nn.Conv2d(64, 128, kernel_size=3)
        self.pool3 = nn.MaxPool2d(2, 2)
        
       # Adaptive pooling layer
        self.adaptive_pool = nn.AdaptiveAvgPool2d((1, 1))
        
        # Fully connected layers
        self.fc1 = nn.Linear(128, num_boxes * 5)  # Adjust input size based on pooling
       # self.fc2 = nn.Linear(num_boxes * 5, num_boxes * 5)
        self.fc3 = nn.Linear(num_boxes * 5, 1)

    def forward(self, x):
         # Convolutional layers with ReLU activation and max pooling
        x = self.pool1(nn.functional.relu(self.conv1(x)))
        x = self.pool2(nn.functional.relu(self.conv2(x)))
        x = self.pool3(nn.functional.relu(self.conv3(x)))
        # Flatten the output for fully connected layers

        x = self.adaptive_pool(x)
        
        x = torch.flatten(x, start_dim=1)
        
        # Fully connected layers with ReLU activation
        x = nn.functional.relu(self.fc1(x))
        #x = nn.functional.relu(self.fc2(x))

        # Output layer
        x = self.fc3(x)
        return x

```

```python
class PNet(nn.Module):
    def __init__(self):
        super(PNet, self).__init__()

        self.channels = [3, 10, 16, 32]
        self.down = nn.Sequential(
            nn.Conv2d(self.channels[0], self.channels[1], kernel_size = 3, stride = 1, padding = False),
            nn.PReLU(self.channels[1]),
            nn.MaxPool2d(kernel_size = 2, stride = 2),
            nn.Conv2d(self.channels[1], self.channels[2], kernel_size = 3, stride = 1, padding = False),
            nn.PReLU(self.channels[2]),
            nn.Conv2d(self.channels[2], self.channels[3], kernel_size = 3, stride = 1, padding = False),
            nn.PReLU(self.channels[3])
        )

        #Classification layer
        self.classification = nn.Sequential(
            nn.Conv2d(self.channels[3], 2, kernel_size = 1, stride = 1, padding = True),
            nn.Softmax(dim = 1)
        )


        #Bounding box regression layer
        self.box = nn.Conv2d(self.channels[3], 4, kernel_size = 1, stride = 1, padding = True)

    def forward(self, x):
        x = self.down(x)
        classification = self.classification(x)
        bounding_box = self.box(x)

        return classification, bounding_box
        
```

```python
class RNet(nn.Module):
    def __init__(self):
        super(RNet, self).__init__()

        self.channels = [3, 28, 48, 64, 128]
        self.down = nn.Sequential(
            nn.Conv2d(self.channels[0], self.channels[1], kernel_size = 3, stride = 1, padding = False),
            nn.PReLU(self.channels[1]),
            nn.MaxPool2d(kernel_size = 3, stride = 2),
            nn.Conv2d(self.channels[1], self.channels[2], kernel_size = 3, stride = 1, padding = False),
            nn.PReLU(self.channels[2]),
            nn.MaxPool2d(kernel_size = 3, stride = 2),
            nn.Conv2d(self.channels[2], self.channels[3], kernel_size = 3, stride = 1, padding = False),
            nn.PReLU(self.channels[3]),
            nn.Flatten(),
            nn.Linear(self.channels[3] * 3 * 3, self.channels[4]),
            nn.PReLU(self.channels[4])
        )

        #Classification layer
        self.classification = nn.Sequential(
            nn.Linear(self.channels[4], 2),
            nn.Softmax(dim = 1)
        )

        #Bounding box regression layer
        self.box = nn.Linear(self.channels[4], 4)
        
    def forward(self, x):
        x = self.down(x)
        classification = self.classification(x)
        bounding_box = self.box(x)

        return classification, bounding_box
```

```python
class ONet(nn.Module):
    def __init__(self):
        super(ONet, self).__init__()

        self.channels = [3, 32, 64, 64, 128, 256]
        self.down = nn.Sequential(
            nn.Conv2d(self.channels[0], self.channels[1], kernel_size = 3, stride = 1, padding = False),
            nn.PReLU(self.channels[1]),
            nn.MaxPool2d(kernel_size = 3, stride = 2),
            nn.Conv2d(self.channels[1], self.channels[2], kernel_size = 3, stride = 1, padding = False),
            nn.PReLU(self.channels[2]),
            nn.MaxPool2d(kernel_size = 3, stride = 2),
            nn.Conv2d(self.channels[2], self.channels[3], kernel_size = 3, stride = 1, padding = False),
            nn.PReLU(self.channels[3]),
            nn.MaxPool2d(kernel_size = 2, stride = 2),
            nn.Conv2d(self.channels[3], self.channels[4], kernel_size = 2, stride = 1, padding = False),
            nn.PReLU(self.channels[4]),
            nn.Flatten(),
            nn.Linear(self.channels[4] * 3 * 3, self.channels[5]),
            nn.PReLU(self.channels[5]),
        )

        #Classification layer
        self.classification = nn.Sequential(
            nn.Linear(self.channels[5], 2),
            nn.Softmax(dim = 1)
        )

        #Bounding box regression layer
        self.box = nn.Linear(self.channels[5], 4)
        
    def forward(self, x):
        x = self.down(x)
        classification = self.classification(x)
        bounding_box = self.box(x)

        return classification, bounding_box
```

```python
class PNetTrainer:
    def __init__(self, model, train_loader, val_loader, lr = 0.001):
        self.device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
        self.model = model.to(self.device)
        self.train_loader = train_loader
        self.val_loader = val_loader
        self.optimizer = optim.RAdam(self.model.parameters(), lr=lr)
        self.classification_loss = nn.CrossEntropyLoss()
        self.box_loss = nn.MSELoss()

    def train(self, epochs):
        for epoch in range(epochs):
            self.model.train()
            train_loss_classification = 0.0
            train_loss_box = 0.0

            for i, (input, labels) in enumerate(self.train_loader):
                input, label = input.to(self.device), labels.to(self.device)

                self.optimizer.zero_grad()

                classification_pred, box_pred = self.model(inputs)
                cls_loss = self.classification_loss(classification_pred, labels[:,0])
                box_loss = self.box_loss(box_pred, labels[:,1:])
                loss = cls_loss + box_loss


                loss.backward()
                self.optimizer.step()

                train_loss_classification += cls_loss.item()
                train_loss_box = box_loss.item()

            train_loss_classification /= len(self.train_loader)
            train_loss_box /= len(self.train_loader)

            print(f"Epoch {epoch+1}/{epochs}, "
                  f"Train Loss_cls: {train_loss_cls:.4f}, "
                  f"Train Loss_bbox: {train_loss_bbox:.4f}")


            self.model.eval()
            val_loss_cls = 0.0
            val_loss_bbox = 0.0
            with torch.no_grad():
                for inputs, labels in self.val_loader:
                    inputs, labels = inputs.to(self.device), labels.to(self.device)

                    cls_pred, bbox_pred = self.model(inputs)
                    cls_loss = self.classification_loss(cls_pred, labels[:, 0])
                    bbox_loss = self.bbox_loss(bbox_pred.squeeze(1), labels[:, 1:])

                    val_loss_cls += cls_loss.item()
                    val_loss_bbox += bbox_loss.item()

                val_loss_cls /= len(self.val_loader)
                val_loss_bbox /= len(self.val_loader)

                print(f"Val Loss_cls: {val_loss_cls:.4f}, "
                      f"Val Loss_bbox: {val_loss_bbox:.4f}")
```
